```python
def character_frequency(word):
    frequencies = {}
    for i in range(len(word) - 1):
        current_char = word[i]
        next_char = word[i + 1]
        if current_char in frequencies:
            if next_char in frequencies[current_char]:
                frequencies[current_char][next_char] += 1
            else:
                frequencies[current_char][next_char] = 1
        else:
            frequencies[current_char] = {next_char: 1}
    return frequencies
```

```python
import string

def count_character_frequency(filename):
    character_frequency = {}
    
    with open(filename, 'r') as file:
        for line in file:
            words = line.split()
            for word in words:
                word = word.lower().strip(string.punctuation)
                
                for i in range(len(word)-1):
                    current_char = word[i]
                    next_char = word[i+1]
                    
                    if current_char not in character_frequency:
                        character_frequency[current_char] = {}
                    
                    if next_char not in character_frequency[current_char]:
                        character_frequency[current_char][next_char] = 0
                    
                    character_frequency[current_char][next_char] += 1
    
    return character_frequency
```

```python
filename = 'englishsample.txt'
EnglishFrequency = count_character_frequency(filename)

for char in EnglishFrequency:
    print(f"Character '{char}':")
    for next_char, count in EnglishFrequency[char].items():
        print(f"  '{next_char}': {count}")
```

```python
sum=0
for char in EnglishFrequency:
    for next_char, count in EnglishFrequency[char].items():
        sum+=count
    
    for next_char, count in EnglishFrequency[char].items():
        EnglishFrequency[char][next_char]=count/sum;
    sum=0

for char in EnglishFrequency:
    print(f"Character '{char}':")
    for next_char, count in EnglishFrequency[char].items():
        print(f"  '{next_char}': {count}")
    
```

```python
def NepNormalize(A):
  import pylcs
  simi=0
  word1=''
  with open('Nepali_Romanized_Dataset.txt', 'r') as file:
          for line in file:
              words = line.split()
              for word in words:
                  word = word.lower().strip(string.punctuation)
                  lcs = pylcs.lcs_sequence_length(A, word)
                  edit = pylcs.edit_distance(A, word)
                  min_len = min(len(A),len(word))
                  sim = lcs/(min_len+edit)
                  if(sim>simi):
                    simi=sim;
                    word1=word
  print(word1) 
```
